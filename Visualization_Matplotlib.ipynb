# Matplotlib

+ The Figure is the overall window or page that everything is drawn on. It’s the top-level component of all.
+ To the figure you add Axes. The Axes is the area on which the data is plotted.
+ A figure can have multiple axes.

<font color='blue'>Note:</font> when you see, for example, plt.xlim, you’ll call ax.set_xlim() behind the covers. 
+ All methods of an Axes object exist as a function in the pyplot module and vice versa. 
+ Mostly, you’ll use the functions of the pyplot module because they’re much cleaner, at least for simple plots!

```python
fig = plt.figure()
# We'll explain the "111" later. Basically, 1 row and 1 column.
ax = fig.add_subplot(111) 
ax.set_xlim([0.5, 4.5])
ax.set_ylim([-2, 8])
ax.text(2.1,3.0,'This is a Subplot',color='red')

ax.set_title('A Different Example Axes Title')
ax.set_ylabel('Y-Axis (changed)')
ax.set_xlabel('X-Axis (changed)')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim(0, 4.5)
ax.set_ylim(5, 35)

# Line Plot
x=[1, 2, 3, 4]
y=[10, 20, 25, 30]
ax.plot(x, y,color='blue', linewidth=6)

#Scatter Plot
x=[0.3, 3.8, 1.2, 2.5]
y=[11, 25, 9, 26]
ax.scatter(x,y, marker='x', color = 'red')

plt.show()
```

## Sub Plots

```python
fig, axes = plt.subplots(2, 2, figsize=(9, 9))
fig.subplots_adjust(wspace=0.5, hspace=0.3)
(ax1,ax2),(ax3,ax4) = axes
ax1.plot([1, 2, 3, 4], [10, 20, 25, 30], color='blue', linewidth=3)
ax2.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], c=[1, 2, 3, 5], marker='^')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(9, 9))
fig.subplots_adjust(wspace=0.3, hspace=0.2)
(ax1,ax2),(ax3,ax4) = axes
ax1.plot([1, 2, 3, 4], [10, 20, 25, 30], color='blue', linewidth=3)
ax2.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], c=[1, 2, 3, 5], marker='^')
ax3.scatter([2, 4, 6, 8, 10], [2, 4, 6, 8, 10], c=[1, 2, 3, 5, 4], marker='x')
ax4.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], c=[6, 7, 9, 11], marker='o')
plt.show()
```

```python
df_used_cars = pd.read_csv("C:/Users/Moin Dalvi/Documents/EXcelR Study and Assignment Material/Data Science/Visualization/CSV files/bar_chart_data.csv")
df_used_cars
```

```python
# Set the size of the chart
plt.figure(figsize = (6, 4)) 
#figsize=plt.figaspect(9/6)


# To create a bar chart with matplotlib you can use the 'bar' function.
plt.bar(x = df_used_cars["Brand"], #specify the x axis
        height = df_used_cars["Cars Listings"], #specify the y axis
        color = "#F5DEB3") # the color for the bars

# rotate and format the labels for the x-axis
plt.xticks( rotation = 90,fontsize = 12) 
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold") 
plt.ylabel("Number of Listings", fontsize = 13 ) 

# you can export your chart as a picture
#plt.savefig("Used Cars Bar.png") 

# depending on which environment you're using without this line you're chart may not display properly.
plt.show() 
```

```python
plt.figure(figsize = (6,4))
plt.bar(df_used_cars['Brand'], df_used_cars['Cars Listings'], color='black')
plt.xticks(rotation = 90, fontsize = 14)
plt.title('Cars Listing by Brand', fontsize = 16, fontweight ='bold')
plt.xlabel("Different Car Brands", fontsize = 14, color = 'red')
plt.ylabel('Numbers of Listings', fontsize = 13)
plt.show()
```

### b. Horizontal Bar Chart

```python
plt.figure(figsize = (6,4))
plt.barh(df_used_cars['Brand'], df_used_cars['Cars Listings'], color='blue')
plt.xticks(rotation = 90, fontsize = 12)
plt.yticks(fontsize=14)
plt.title('Cars Listing by Brand', fontsize = 16, fontweight ='bold')
plt.xlabel("Numbers of Listings", fontsize = 14, color = 'red')
plt.ylabel('Different Car Brands', fontsize = 13, color = 'red' )
plt.show()
```

# Input Data: Bivariate (2D Series)

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]


# create a line chart, years on x-axis, gdp on y-axis
#plt.plot(years, gdp, color='green', marker='x', linestyle='dashed')
plt.plot(gdp, years, color='midnightblue', marker='o', linestyle='solid')


# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.xlabel("Year")

plt.show()
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.plot(years, gdp,  color = 'black', marker ='X', linestyle = 'solid')
plt.title('Nominal GDP')
plt.xlabel('Year')
plt.ylabel('Billions of $')
sn.despine(right = True, top = True)
plt.show()
```
